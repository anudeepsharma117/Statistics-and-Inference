Answer 1 - Type I Error (False Positive)

It happens when we reject the null hypothesis even though it is actually true.
Meaning: We think something happened, but in reality nothing happened.

Example: Test says a person is sick, but they are actually healthy.

Type II Error (False Negative)

It happens when we fail to reject the null hypothesis even though it is false.
Meaning: Something actually happened, but we failed to notice it.

Example: Test says a person is healthy, but they actually have a disease.

Impact on decision-making:

Type I error: Leads to taking wrong actions.

Type II error: Leads to missing important actions.

Question 2:What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?


A P-value is the probability of getting the observed results (or more extreme results) if the null hypothesis is actually true.
. If P-value is small (usually < 0.05):

The result is unlikely to be due to chance.

We reject the null hypothesis.

If P-value is large (≥ 0.05):

The result could easily happen by chance.

We fail to reject the null hypothesis.
. Low P-value → Evidence against H₀ → Reject H₀

High P-value → Not enough evidence → Do not reject H₀

Question 3:Explain the difference between a Z-test and a T-test, including when to use
each.


Z-test is used when the sample size is large (n > 30).

T-test is used when the sample size is small (n ≤ 30).

Z-test is used when the population standard deviation (σ) is known.

T-test is used when the population standard deviation is unknown, and we use the sample standard deviation instead.

Z-test assumes the data follows a normal distribution.

T-test is more flexible and can be used even if data is not perfectly normal, especially with small samples.  

  *Use Z-test:

Large sample size

Population standard deviation known

Use T-test:

Small sample size

Population standard deviation unknown

Question 4:What is a confidence interval, and how does the margin of error influence
its width and interpretation?


Confidence Interval (CI)

A confidence interval is a range of values that is likely to contain the true population value.

It is written as:
Estimate ± Margin of Error

Margin of Error (MOE) and Its Influence

The margin of error tells us how much we allow our estimate to vary.

Large margin of error → Wider confidence interval

Means we are less precise.

Small margin of error → Narrow confidence interval

Means we are more precise.
A wider CI means more uncertainty about the true value.

A narrow CI means more confidence and accuracy in the estimate.

Question 5: Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?


ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others. It helps avoid doing multiple t-tests, which can increase the chance of errors. ANOVA works under three main assumptions: the data in each group should be normally distributed, all groups should have similar variances, and the observations must be independent. It extends hypothesis testing by allowing the comparison of more than two groups at the same time, instead of comparing groups pair-by-pair like a t-test.

Question 6: Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.


```python
import numpy as np
from scipy.stats import norm

# Sample data
data = [50, 52, 49, 48, 51, 53, 47, 50, 52, 49]  # Example dataset

# Known population mean (H0)
population_mean = 50

# Known population standard deviation (σ)
population_std = 3

# Calculate sample mean
sample_mean = np.mean(data)

# Sample size
n = len(data)

# Z-test statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# P-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print("Sample Mean:", sample_mean)
print("Z-Score:", z_score)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference found.")
```

Question 8: Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a non-normal distribution (Exponential)
population = np.random.exponential(scale=1, size=50000)

# Step 2: Draw multiple samples and compute their means
sample_means = []
num_samples = 1000       # number of samples
sample_size = 30         # sample size for each sample

for _ in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

# Step 3: Plot histogram of original non-normal data
plt.hist(population, bins=40)
plt.title("Original Non-Normal Distribution (Exponential)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

# Step 4: Plot histogram of sample means (CLT)
plt.hist(sample_means, bins=30)
plt.title("Distribution of Sample Means (Central Limit Theorem)")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.show()
```

Question 9: Write a Python function to calculate and visualize the confidence interval
for a sample mean.


```python
# Install scipy if needed
!pip install scipy

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_confidence_interval(data, confidence=0.95):
    # Calculate statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)

    # Z-value for the given confidence level
    z_value = norm.ppf((1 + confidence) / 2)

    # Margin of error
    margin_error = z_value * (sample_std / np.sqrt(n))

    # Confidence interval
    lower = sample_mean - margin_error
    upper = sample_mean + margin_error

    # Print CI
    print("Sample Mean:", sample_mean)
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower}, {upper})")

    # Clear previous plots
    plt.figure(figsize=(8, 5))

    # Plot histogram
    plt.hist(data, bins=20, alpha=0.6)

    # Plot CI lines
    plt.axvline(sample_mean, linestyle="solid", label="Sample Mean")
    plt.axvline(lower, linestyle="--", label="Lower CI")
    plt.axvline(upper, linestyle="--", label="Upper CI")

    plt.title("Confidence Interval Visualization")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()

# Example usage
data = np.random.normal(loc=50, scale=10, size=100)
plot_confidence_interval(data)
```

Question 10: Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.


```python
# Install scipy if not installed
!pip install scipy

import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example)
observed = np.array([25, 30, 20, 25])

# Expected frequencies (example - uniform distribution)
expected = np.array([25, 25, 25, 25])

# Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed distribution is different from expected.")
else:
    print("Fail to reject the null hypothesis: No significant difference found.")
```
